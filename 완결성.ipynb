```python
import pandas as pd
```

```python
df = pd.read_csv('/content/attendance.csv', index_col=0)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
# 결측값을 찾아라~~
```

```python
# 결측값이 있을때 좋은 데이터가 있고 아닌데이터가 있다 구분하자
df.dropna() # 이건 결측값을 지워주지만 저장은 안한다
# df.dropna(axis='culumns') 는 결측값이 있는 컬럼을 없앤다.
```

```python
df.dropna(inplace=True)
# 새로운 데이터를 만들기 위해 인플레이스 추가
```

```python
#위 같은 경우는 년도가 날라가버린다.
# 결측값을 대체하는 방법이 좋은 방법일 수도?
```

```python
df = pd.read_csv('/content/attendance.csv', index_col=0)
df
```

```python
df.fillna(0) # 결측값을 0으로 대체
```

```python
df.mean()
```

```python
df.fillna(df.mean())#결측값을 평균값으로 대체
```

```python
df.fillna(df.median()) #이건 중간값으로 대체
# 영구적으로 데이터셋을 만들고 싶다면 inplace=True
```
