# 정확성

```python
# 이상점 다른 데이터와 너무 차이나는 데이터
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('/content/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
#데이터에서 이상점 지우기 단, 항상 지우는 건 아님
df['abv'].describe() # 퍼센트 지점 확인
```

```python
q1 = df['abv'].quantile(0.25) #25퍼 지점을 받아오
q1
```

```python
q3 = df['abv'].quantile(0.75)#75퍼 지점을 받아오
q3
```

```python
iqr = q3 - q1
```

```python
iqr #이걸로 이상점을 찾을수 있다?
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
# 25퍼보다 밑에 있거나 75퍼보다 위에 있거
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055 # 잘못입력한 맥주값은 변경해주고
```

```python
df.loc[2250]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace=True) # 맥주아닌것들 날렷!
df[condition]
```

```python
df.plot(kind='box', y='abv')
```
